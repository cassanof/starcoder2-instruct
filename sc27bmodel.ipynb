````python
from code_editing.models import DirectEditModel, starcoder2_edit_prompt_1shot, python_markdown_codeblock_extract
````

````python
def wrap_prompt(x, y, z):
    res = starcoder2_edit_prompt_1shot(x, y, z)
    print(res)
    return res

def wrap_post(x, y):
    y = "```py\n" + y
    res = python_markdown_codeblock_extract(x, y)
    print("###### OUTS #########")
    print(y)
    print("###### POST PROCESSED #########")
    print(res)
    return res

model = DirectEditModel(
            "/scratch/models/sc2-7b-base/",
            completion_engine="transformers",
            num_gpus=1,
            prompt_format=wrap_prompt,
            post_process=wrap_post,
        )
````

````python
import datasets
ds = datasets.load_dataset("nuprl-staging/CanItEdit", split="test")
````

````python
ex = ds[0]
before = ex["before"]
ins = ex["instruction_lazy"]
p = {"content": before, "instruction": ins}
````

````python
model.generate([p])
````
